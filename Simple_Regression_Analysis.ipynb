```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv("/content/Study Hour(Linear Regression).csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
x=dataset[["Hours"]]
y=dataset[["Scores"]]
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="Hours",y="Scores",data=dataset)
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
#y=m*x+c
#y=9.74054996*x	+1.7557382
#y
```

```python
lr.coef_  # from the value we will found that angle is between 0 to 90
             #if value is +ve then angle is between 0 to 90
             #if value is -ve then angle is grater than  90
```

```python
lr.intercept_
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="Hours", y="Scores", data=dataset)
y_pred = lr.predict(x)
plt.plot(x, y_pred, c="red", label="Best Fit Line")

plt.legend()
plt.show()
```
