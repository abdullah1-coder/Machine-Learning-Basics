**Data Cleaning**

Handling Missing Values

```python
dataset.fillna(dataset.mode())
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

Handling Duplicates

```python
dataset.duplicated().sum()
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
dataset.duplicated().sum()
```

Converting Datatypes

```python
dataset = dataset[dataset['Inches'] != '?']
dataset['Inches_Float'] = dataset['Inches'].astype(float)
```

```python
dataset.info()
```

Renaming or dropping irrelevant columns

```python
dataset = dataset.drop("Inches", axis=1)
```

```python
dataset.columns
```

```python
dataset.rename(columns={"OpSys":"Operating System"},inplace=True)
```

```python
dataset.columns
```

**Exploratory Data Analysis (EDA)**

Univariate Analysis
*Histogram*

```python
plt.figure(figsize=(10, 6))
sns.histplot(dataset['Price'], kde=True, color='blue')
plt.title('Distribution of Laptop Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

Bivariate Analysis *Box Plot*

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x='Company', y='Price', data=dataset)
plt.xticks(rotation=45) # Rotate labels so they don't overlap
plt.title('Price Distribution by Brand')
plt.show()
```

Bivariate Analysis *Correlation Heatmap*

```python
plt.figure(figsize=(10, 8))
# Select only numeric columns for correlation
numeric_df = dataset.select_dtypes(include=['float64', 'int64'])
corr = numeric_df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

Scatter Plot: *RAM vs. Price*

```python

dataset['Ram'] = dataset['Ram'].astype(str).str.replace('GB', '').astype(int)


plt.figure(figsize=(10, 6))
sns.scatterplot(x='Ram', y='Price', data=dataset, hue='TypeName', s=100)
plt.title('Relationship between RAM and Price')
plt.xlabel('RAM (GB)')
plt.ylabel('Price')
plt.legend(title='Laptop Type')
plt.show()
```

Count Plot: *Number of Laptops by Brand*

```python

plt.figure(figsize=(12, 6))

order = dataset['Company'].value_counts().index
sns.countplot(y=dataset['Company'], order=order, palette='viridis')
plt.title('Number of Laptops by Brand (Market Share)')
plt.xlabel('Count')
plt.ylabel('Brand')
plt.show()
```

**Business Insights & Interpretation**:  
After analyzing the data, here are the key takeaways regarding the laptop market:

1. **Most Laptops Target Budget-Conscious Buyers**

 The price histogram is heavily skewed to the right. This means the vast majority of laptops sit in the affordable to mid-range price bracket (around 20kâ€“80k).

 While you do see some incredibly expensive machines, they are rare outliers. The bulk of the market is fighting for the average consumer who is looking for value. Manufacturers have to compete aggressively on price to win in this crowded space, whereas the high-end market is less about price and more about specific performance needs or luxury status.

2. **Apple and Razer Stick to the High End**

 When looking at the box plots for each company, Razer and Apple stand out with the highest median prices. Their price ranges are also quite narrow compared to others.

 These brands have a clear "premium" strategy. You don't really see "budget" Apple or Razer laptops. They cater to a specific audienceâ€”gamers and creativesâ€”who are willing to pay more for brand reputation and build quality. LG also surprisingly sits high up, likely due to their specific focus on high-end ultrabooks like the Gram.

3. **The Big Players Cover Every Base**

Giants like Dell, HP, Lenovo, and Asus have massive price ranges. Their box plots stretch from very low to very high prices.

These companies are trying to be everything to everyone. They have cheap, entry-level notebooks for students, but they also sell top-tier gaming rigs and workstations. This "mass market" approach helps them capture the largest market share, but it also means they have to manage a very complex lineup of products.

4. **The Race to the Bottom**

 Smaller brands like Vero, Chuwi, and Mediacom have very low median prices and almost no price variation.

 These brands are competing almost entirely on being the cheapest option available. They target customers for whom price is the only factor that matters. To keep prices this low, they likely operate on thin margins and focus on basic functionality over performance or durability.

5. **You Pay for Power, Not Size**

The correlation heatmap shows that specs like RAM have a much stronger relationship with price than screen size does.

 A bigger screen doesn't necessarily mean a more expensive laptopâ€”there are plenty of cheap 15-inch models. However, bumping up the RAM or getting a better CPU almost always drives the price up significantly. This suggests that customers are paying for what's inside the chassis (performance) rather than just the physical size of the device.
