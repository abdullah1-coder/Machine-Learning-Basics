```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("/content/SalaryMulti.csv")
dataset.head(3)
```

```python
plt.figure(figsize=(5,4))
sns.pairplot(data=dataset)
plt.show()
```

```python
sns.heatmap(data=dataset.corr(),annot=True)
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.feature_names_in_
```

```python
# y=2988.17486511*Total Experience+1898.17107701*Team Lead Experience+974.70334101*Project Manager Experience+371.32169738*Certifications + 50769.345403297055
```

```python
lr.score(x_test,y_test)
```

```python
lr.predict([[7,2,4,1]])
```

```python
import numpy as np

# Get input from user
total_exp = float(input("Enter Total Experience (in years): "))
team_lead_exp = float(input("Enter Team Lead Experience (in years): "))
pm_exp = float(input("Enter Project Manager Experience (in years): "))
certifications = int(input("Enter number of Certifications: "))

# Make prediction
features = np.array([[total_exp, team_lead_exp, pm_exp, certifications]])
prediction = lr.predict(features)

print(f"Predicted Salary: {prediction[0]:,.2f}")
```
