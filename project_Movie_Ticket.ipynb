```python
import pandas as pd

MOVIE_FILE = "/content/@0 movies.xlsx"

def load_movies():
    try:
        df = pd.read_excel(MOVIE_FILE)
    except FileNotFoundError:
        df = pd.DataFrame(columns=["MovieID", "MovieName", "TicketPrice", "AvailableSeats", "ShowTime"])
        df.to_excel(MOVIE_FILE, index=False)
    return df
```

```python
load_movies()
```

```python
def add_movie():
    df = load_movies()

    movie_id = input("Enter Movie ID: ")
    movie_name = input("Enter Movie Name: ")
    price = float(input("Enter Ticket Price: "))
    seats = int(input("Enter Available Seats: "))
    show_time = input("Enter Show Timings: ")

    new_row = {
        "MovieID": movie_id,
        "MovieName": movie_name,
        "TicketPrice": price,
        "AvailableSeats": seats,
        "ShowTime": show_time
    }

    df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
    df.to_excel(MOVIE_FILE, index=False)

    print("Movie added successfully!")
```

```python
def view_movies():
    try:
        df = pd.read_excel(MOVIE_FILE)
    except FileNotFoundError:
        print("No movies found. Please add some movies first.")
        return

    if df.empty:
        print("No movies available in the file.")
        return

    print("\n--- Available Movies ---\n")

    for index, row in df.iterrows():
        print(f"Movie ID: {row['MovieID']}")
        print(f"Movie Name: {row['MovieName']}")
        print(f"Ticket Price: {row['TicketPrice']}")
        print(f"Available Seats: {row['AvailableSeats']}")
        print(f"Show Time: {row['ShowTime']}")
        print("-" * 30)
```

```python
view_movies()
```

```python


BOOKINGS_FILE = "/content/bookings.xlsx"

def book_ticket():
    # Load movies file
    try:
        movies_df = pd.read_excel(MOVIE_FILE)
    except FileNotFoundError:
        print("No movies found. Please add some movies first.")
        return

    if movies_df.empty:
        print("No movies available.")
        return

    # Show movies to user
    print("\n--- Available Movies ---\n")
    print(movies_df[["MovieID", "MovieName", "TicketPrice", "AvailableSeats", "ShowTime"]])
    print()

    # Select movie
    movie_id = input("Enter the Movie ID you want to book: ")

    if movie_id not in movies_df["MovieID"].astype(str).values:
        print("Invalid Movie ID!")
        return

    movie = movies_df[movies_df["MovieID"].astype(str) == movie_id].iloc[0]

    # Ask for number of tickets
    num_tickets = int(input("Enter number of tickets: "))

    if num_tickets > movie["AvailableSeats"]:
        print("Not enough seats available!")
        return

    # Calculate total price
    total_price = num_tickets * movie["TicketPrice"]

    # Update available seats
    movies_df.loc[movies_df["MovieID"].astype(str) == movie_id, "AvailableSeats"] -= num_tickets
    movies_df.to_excel(MOVIE_FILE, index=False)

    # Save booking
    try:
        booking_df = pd.read_excel(BOOKINGS_FILE)
    except FileNotFoundError:
        booking_df = pd.DataFrame(columns=["MovieID", "MovieName", "NumTickets", "TotalPrice"])

    new_booking = {
        "MovieID": movie_id,
        "MovieName": movie["MovieName"],
        "NumTickets": num_tickets,
        "TotalPrice": total_price
    }

    booking_df = pd.concat([booking_df, pd.DataFrame([new_booking])], ignore_index=True)
    booking_df.to_excel(BOOKINGS_FILE, index=False)

    # Final summary
    print("\n--- Booking Successful! ---")
    print(f"Movie: {movie['MovieName']}")
    print(f"Tickets Booked: {num_tickets}")
    print(f"Total Price: {total_price}")
    print("----------------------------")
```

```python
book_ticket()
```

```python
def search_movie():
    df = pd.read_excel(MOVIE_FILE)

    query = input("Enter Movie ID or Name to search: ").lower()

    # handle MovieID search
    if query.isdigit():
        results = df[df['MovieID'] == int(query)]
    else:
        results = df[df['MovieName'].str.lower().str.contains(query, na=False)]

    if results.empty:
        print("\nNo movie found.")
    else:
        print("\n--- Search Results ---\n")
        print(results)

```

```python
BOOKING_FILE = "/content/bookings.xlsx"   # or your actual file

def view_bookings():
    try:
        df = pd.read_excel(BOOKING_FILE)

        if df.empty:
            print("\nNo bookings found yet.")
            return

        print("\n--- All Bookings ---\n")
        print(df.to_string(index=False))   # cleaner table view

    except FileNotFoundError:
        print("\nNo bookings found yet. File does not exist!")
```

```python
view_bookings()
```

```python
def main_menu():
    while True:
        print("\n===== MOVIE TICKETING SYSTEM =====")
        print("1. Add Movie")
        print("2. View Movies")
        print("3. Book Ticket")
        print("4. Search Movie")
        print("5. View Bookings")
        print("6. Exit")



        choice = input("\nEnter your choice: ")

        if choice == "1":
            add_movie()

        elif choice == "2":
            view_movies()

        elif choice == "3":
            book_ticket()

        elif choice == "4":
            search_movie()

        elif choice == "5":
            view_bookings()

        elif choice == "6":
            print("Exiting program...")
            break

        else:
            print("Invalid choice! Please enter 16.")
```
